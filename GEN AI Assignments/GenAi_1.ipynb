```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Generate synthetic data
np.random.seed(42)
num_samples = 300

g1 = np.random.normal(loc=2, scale=1, size=(num_samples, 2))  # Cluster 1
g2 = np.random.normal(loc=-2, scale=1, size=(num_samples, 2)) # Cluster 2
g3 = np.random.normal(loc=5, scale=1.5, size=(num_samples, 2)) # Cluster 3

data = np.vstack((g1, g2, g3))  # Combine clusters

# Fit Gaussian Mixture Model
num_components = 3
gmm = GaussianMixture(n_components=num_components, covariance_type='full', random_state=42)
gmm.fit(data)
labels = gmm.predict(data)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=10, alpha=0.7)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100, label='Cluster Centers')
plt.title("Gaussian Mixture Model Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
