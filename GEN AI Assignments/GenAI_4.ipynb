```python
from transformers import pipeline

# Load a pre-trained question-answering pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Define a sample context (Modify as needed)
context = """
Artificial Intelligence (AI) is a branch of computer science that aims to create machines that can perform tasks that typically require human intelligence.
These tasks include learning, reasoning, problem-solving, perception, language understanding, and more.
Modern AI uses techniques such as machine learning and deep learning to process large amounts of data efficiently.
"""

# Chatbot loop
print("QA Chatbot is ready! Type 'exit' to quit.")
while True:
    question = input("\nAsk a question: ")
    if question.lower() == "exit":
        print("Goodbye!")
        break

    # Get the answer from the model
    result = qa_pipeline(question=question, context=context)

    # Print the answer
    print("Answer:", result["answer"])
```

```python
from google.colab import userdata
import google.generativeai as genai

key = userdata.get('GeminiKey')
genai.configure(api_key=key)
model = genai.GenerativeModel('gemini-1.5-flash')
```

```python
def chatbot():
  print("Hello! 👋 How can I assist you today? Ask me anything!(Type 'Exit/Quit/Finish' to end the conversation!)\n")

  while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit' or user_input.lower() == 'quit' or user_input.lower() == 'finish':
      print("Goodbye! 👋")
      break
    response = model.generate_content(user_input)
    print("Bot: " + response.text)

chatbot()
```
