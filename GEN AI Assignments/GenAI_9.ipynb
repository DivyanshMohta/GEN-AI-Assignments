```python

from diffusers import StableDiffusionPipeline
import torch
```

```python

# Load pre-trained model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Use GPU for faster processing
```

```python
import os
from PIL import Image
```

```python

# Create output folder
os.makedirs("frames", exist_ok=True)
```

```python

# Generate multiple frames
prompt = "A dramatic steampunk airship sailing above a Victorian-era city, with intricate mechanical details, billowing steam clouds, and a richly detailed, sunset sky as the backdrop."
num_frames = 10  # Number of frames
```

```python

import imageio
```

```python
frames = []
for i in range(num_frames):
    frame = Image.open(f"frames/frame_{i:03d}.png")
    frames.append(frame)
```

```python

# Save as a video
output_video = "animation.mp4"
imageio.mimsave(output_video, frames, fps=10)  # 10 FPS

print(f"Video saved as {output_video}")
```
